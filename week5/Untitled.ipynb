```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\Fujitsu T\Desktop\uni\week5\images.jfif')
image2= cv2.imread(r'C:\Users\Fujitsu T\Desktop\uni\week5\img2.webp')
plt.imshow(image)
```

# Task 1

## adjust brightness

```python
value = np.clip(150, -255, 255)
brightness_matrix = np.full(image.shape, (value, value, value), dtype=np.uint8)
if value >= 0:
        new_image = cv2.add(image, brightness_matrix)
        plt.imshow(new_image)
else:
        new_img = cv2.subtract(image, np.abs(brightness_matrix))
        plt.imshow(new_image)
```

## contrast adjustment

```python
alpha=2.0
beta=30
new_img = cv2.convertScaleAbs(image, alpha, beta)
plt.imshow(new_img)
```

## gamma correction

```python
gamma_val=2.0
inv_gamma = 1.0 / gamma_val
corrected = np.uint8(np.power(image / 255.0, inv_gamma) * 255.0)
plt.imshow(corrected)
```

## inverse transformation

```python
inv_img=255-image
plt.imshow(inv_img)
```

```python
foreground_img=cv2.resize(foreground_img, (image2.shape[1], image2.shape[0]))
foreground2 = cv2.cvtColor(foreground_img, cv2.COLOR_BGR2RGB)
lower_green = np.array([35, 100, 100])
upper_green = np.array([85, 255, 255])
mask = cv2.inRange(foreground2, lower_green, upper_green)
mask_inv = cv2.bitwise_not(mask)
fg_part = cv2.bitwise_and(foreground_img, foreground_img,mask_inv)
bg_part = cv2.bitwise_and(image2, image2,mask)
new_image = cv2.add(fg_part, bg_part)
plt.imshow(new_image)
```
